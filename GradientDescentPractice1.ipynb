```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
data = pd.read_csv('data.csv')
```

```python
X = data.iloc[:, 0]
```

```python
X
```

```python
Y = data.iloc[:, 1]
```

```python
Y
```

```python
plt.scatter(X, Y)
```

```python
plt.show()
```

```python
m = 0
c = 0
learn_rate = 0.0001
iteration = 1000
n = len(X)
prev_cost = 0
for i in range(iteration):
    y_pred = m * X + c
    mse = (1/n) * sum([val**2 for val in (Y - y_pred)])
    der_m = (-2/n)*sum(X*(Y - y_pred))
    der_c = (-2/n)*sum((Y - y_pred))
    m = m - der_m * learn_rate
    c = c - der_c * learn_rate
    if math.isclose(mse, prev_cost, rel_tol=1e-20):
        break
    prev_cost = mse
print(m , c, mse)
```

```python
y_pred = m * X + c
plt.scatter(X,Y)
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')
plt.show()
```
